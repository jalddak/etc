```python
rating_item_sort_file = 'rating_data_item_sort.csv'
rating_item_sort_data = pd.read_csv(os.path.join(basicpath, rating_item_sort_file), names=['user_id', 'item_id', 'rating'])
rating_item_sort_data
```

```python
meta_sort_file = 'meta_data_sort.csv'
meta_sort_data = pd.read_csv(os.path.join(basicpath, meta_sort_file), names=['item_id', 'title', 'brand', 'category', 'imageURL', 'imageURLHighRes'])
meta_sort_data
```

```python
meta_sort_file = 'meta_data_sort.json'
meta_sort_data = pd.read_json(os.path.join(basicpath, meta_sort_file))
meta_sort_data
```

```python
rating_del_not_in_meta = rating_item_sort_data[rating_item_sort_data['item_id'].isin(meta_sort_data['item_id'])]
```

```python
rating_del_not_in_meta = rating_del_not_in_meta.sort_values('user_id')
rating_del_not_in_meta = rating_del_not_in_meta.reset_index()
del rating_del_not_in_meta['index']
rating_del_not_in_meta
```

```python
duplication_count = rating_del_not_in_meta.groupby('user_id').count()
```

```python
condition = duplication_count.item_id >= 5
more_than_five_reviews = duplication_count[condition]
more_than_five_reviews = more_than_five_reviews.reset_index()
more_than_five_reviews
```

```python
rating_mtf = rating_del_not_in_meta[rating_del_not_in_meta['user_id'].isin(more_than_five_reviews['user_id'])]
rating_mtf = rating_mtf.reset_index()
del rating_mtf['index']
rating_mtf
```

```python
rating_mtf.to_csv(os.path.join(basicpath, 'rating_data_end.csv'), header= None)
```

```python
rating_end_file = 'rating_data_end.csv'
rating_end_data = pd.read_csv(os.path.join(basicpath, rating_end_file), names=['user_id', 'item_id', 'rating'])
rating_end_data
```

```python
meta_del_not_in_rating = meta_sort_data[meta_sort_data['item_id'].isin(rating_mtf['item_id'])]
meta_del_not_in_rating = meta_del_not_in_rating.reset_index()
del meta_del_not_in_rating['index']
meta_del_not_in_rating
```

```python
meta_del_not_in_rating.to_csv(os.path.join(basicpath, 'meta_data_end.csv'), header= None)
```

```python
meta_del_not_in_rating.to_json(os.path.join(basicpath, 'meta_data_end.json'), orient = 'records')
```
