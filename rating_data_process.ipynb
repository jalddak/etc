**attribute의 의미 및 종류 / 그리고 삭제할 attribute 선정 및 기준 (preprocessing)**

유저의 아이템에 대한 선호도를 바탕으로 collaborating filtering 방식으로, matrix factorization 을 사용할 것이기 때문에 
기본적으로는 **userid, itemid, rating** 만 반영되면 된다.

---

**overall : 평점을 의미**

verified : 의미를 알 수 없다 -> 삭제

vote : 해당 리뷰에 공감한다는 표시같음.

reviewTime : 리뷰를 남긴 시각 -> 시간에 따른 아이템의 선호도를 확인하여 최근의 선호도를 반영한다면 좋을 수 있겠지만, 아직 거기까지 생각하는 것은 무리 -> 삭제

**reviewerID : 리뷰를 남긴 사람을 의미**

**asin : 물건의 고유번호**

style - format : 책의 재료? 를 말하는 것 같으나, 여기선 쓸데없다고 판단했고, 또한 데이터가 없는 것도 있었기 때문에 -> 삭제

reviewerName : reviewerID 와 동일한의미, 그리고 중복가능성 -> 삭제

reviewText : 리뷰내용, 필요없다 -> 삭제

summary : 한줄요약 -> 삭제

unixReviewTime : 리뷰남긴 시간 unix형식 -> 삭제

```python
output = open(os.path.join(basicpath, 'rating_data_delete_not_need_attribute.csv'), 'w', encoding='utf-8')
with open(os.path.join(basicpath, five_core_review_data_file), "r", encoding="utf-8") as json_file:
    json_string = json_file.readline()
    while json_string:
        data = json.loads(json_string)
        output.write(str(data['reviewerID']) + ',')
        output.write(str(data['asin']) + ',')
        output.write(str(data['overall']) + '\n')
        json_string = json_file.readline()
output.close()
```
